```python
from IPython.display import display
#from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

#%matplotlib nbagg
#%matplotlib inline
%matplotlib widget
```

```python
x = np.linspace(0,2,1000)
fig, ax = plt.subplots(1, figsize=(10, 4))
plt.suptitle('Sine wave')

#amp= 1
#freq = 5
#phase= 1


def update_plot(amp, phase, freq):
    ax.clear()
    y = amp * np.sin(2 * np.pi * (freq * x + phase))
    #display(y)
    #y = x * amp + phase - freq
    ax.plot(x,y)
    plt.show()
#update_plot(amp, phase, freq)
amp = widgets.FloatSlider(min=1, max=10, value=4, description='Amp')
phase = widgets.FloatSlider(min=0, max=5, value= 0, description='Phase')
freq = widgets.FloatSlider(min=1, max=10, value=1, description='Freq')

widgets.interact(update_plot, amp = amp, phase = phase, freq = freq)

#interactive(update_plot, amp = (1,10), phase = (0,5), freq = (1,10))
```
